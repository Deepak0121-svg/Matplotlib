```python
#Creating labels for plot
x=np.array([1,2,3,4,5])
y=np.array([3000,4500,1250,1500,3500])


plt.plot(x,y)
plt.xlabel('Id')
plt.ylabel('salary')
plt.show()
```

```python
plt.plot(x,y,marker='^',mfc='r')
plt.title("Employee salary detail")
plt.xlabel('Id')
plt.ylabel('salary')
plt.show()
```

```python
font1 = {'family':'serif','color':'Black','size':15}
font2 = {'family':'serif','color':'red','size':10}
plt.plot(x,y,marker='^',mfc='r')
plt.title("Employee salary detail",fontdict=font1)
plt.xlabel('Id',fontdict=font2)
plt.ylabel('salary',fontdict=font2)
plt.show()
```

```python
plt.plot(x,y,marker='^',mfc='r')
plt.title("Employee salary detail",fontdict=font1,loc='left')
plt.xlabel('Id',fontdict=font2)
plt.ylabel('salary',fontdict=font2)
plt.show()
```

```python
#matgrid
plt.plot(x,y,marker='^',mfc='r')
plt.title("Employee salary detail",fontdict=font1)
plt.xlabel('Id',fontdict=font2)
plt.ylabel('salary',fontdict=font2)
plt.grid()
plt.show()
```

```python
plt.plot(x,y,marker='^',mfc='r')
plt.title("Employee salary detail",fontdict=font1)
plt.xlabel('Id',fontdict=font2)
plt.ylabel('salary',fontdict=font2)
plt.grid(axis= 'x')
plt.show()
```

```python
plt.plot(x,y,marker='^',mfc='r')
plt.title("Employee salary detail",fontdict=font1)
plt.xlabel('Id',fontdict=font2)
plt.ylabel('salary',fontdict=font2)
plt.grid(axis= 'y')
plt.show()
```

```python
plt.plot(x,y,marker='p',mfc='r')
plt.title("Employee salary detail",fontdict=font1)
plt.xlabel('Id',fontdict=font2)
plt.ylabel('salary',fontdict=font2)
plt.grid(axis= 'x',color='red',ls='-.',lw=0.8)
plt.grid(axis= 'y',color='c',ls=':',lw=0.9)

plt.show()
```

```python
#subplot
x=np.array([1,3,6,7])
y=np.array([10,15,19,20])
plt.subplot(1,3,1)
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('Firstsubplot',fontdict=font1)
plt.show()
plt.subplot(2,2,1)
plt.title('secondsubplot',fontdict=font1,color='c')
plt.plot(x,y)
plt.show()
plt.subplot(3,2,1)
plt.plot(x,y)
plt.title('Thirdsubplot',fontdict=font1,color='r')
plt.show()
```

```python
df=pd.read_csv('ilk5.csv')
df
```

```python
#scatter
x=df['Country']
y=df['Population']
plt.scatter(x,y,color='r')
plt.title('Population scatter chart')
plt.show()
```

```python
x=df['Country']
y=df['Population']
plt.scatter(x,y,color='m')


x=df['Country']
y=df['Revenue']
plt.scatter(x,y,color='c')

plt.show()
```

```python
x=df['Country']
y=df['Population']
colors=np.array(['Green','yellow','red','c','hotpink'])



plt.scatter(x,y,color=colors)
plt.colorbar()

plt.show()
```

```python
#colorMap
x=df['Country']
y=df['Population']
colors=np.array([0,10,20,30,50])



plt.scatter(x,y,c=colors,cmap='hot')

plt.show()
```

```python
x=df['Country']
y=df['Population']
colors=np.array([0,10,20,30,50])



plt.scatter(x,y,c=colors,cmap='Pastel2')
plt.colorbar()

plt.show()
```

```python
x=df['Country']
y=df['Revenue']
colors=np.array([0,10,20,30,50])



plt.scatter(x,y,c=colors,cmap='rainbow')
plt.colorbar()

plt.show()
```

```python
#Matplot bar
x=df['Country']
y=df['Revenue']
font3={'family': 'cursive', 'color': 'r', 'weight': 'light', 'size': 20}

plt.bar(x,y,color='b')
plt.title('Revenue of each country',fontdict=font3)
plt.show()
```

```python
x=df['Country']
y=df['Population']
font3={'family': 'cursive', 'color': 'g', 'weight': 'light', 'size': 20}

plt.bar(x,y,color='c')
plt.title('Population of each country',fontdict=font3)
plt.show()
```

```python
#Horizontal bar
plt.barh(x,y,color='c')
plt.title('Population of each country',fontdict=font3,loc='left')
plt.show()
```

```python
plt.bar(x,y,color='c',width=0.2)
plt.title('Population of each country',fontdict=font3)
plt.show()
```
