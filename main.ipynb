```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Matplotlip ploting
x=np.array([10,20,30,50])
y=np.array([1,2,3,4])

plt.plot(x,y)
plt.show()
```

```python
x=np.array([3,6,9,12])
y=np.array([9,12,18,6])

plt.plot(x,y)
plt.show()
```

```python
y=np.array([12,18,3,6])
plt.plot(y)
plt.show()
```

```python
#Matplotlip marker
y=np.array([13,5,19,2])
plt.plot(y,marker='D')
plt.show()
```

```python
a=np.array([13,2,0,7])
plt.plot(a,'d-.k')
plt.show()
```

```python
#Marker size
plt.plot(a,marker='d',ms=15)
plt.show()
```

```python
a=np.array([1,10,2,14])
plt.plot(a,marker='h',ms=16,mec='m') #markeredgecolor
plt.show()
```

```python
#Markerfacecolor 
a=np.array([1,10,2,14])
plt.plot(a,marker='H',ms=16,mfc='hotpink') 
plt.show()
```

```python
#Matplotlip line
a=np.array([1,10,2,14])
plt.plot(a,ls='-.',color='hotpink')
plt.show()
```

```python
#linestyle
#':','--','-.'
```

```python
plt.plot(a,c='m')
```

```python
#Creating labels for plot
x=np.array([1,2,3,4,5])
y=np.array([3000,4500,1250,1500,3500])


plt.plot(x,y)
plt.xlabel('Id')
plt.ylabel('salary')
plt.show()
```

```python
plt.plot(x,y,marker='^',mfc='r')
plt.title("Employee salary detail")
plt.xlabel('Id')
plt.ylabel('salary')
plt.show()
```

```python
font1 = {'family':'serif','color':'Black','size':15}
font2 = {'family':'serif','color':'red','size':10}
plt.plot(x,y,marker='^',mfc='r')
plt.title("Employee salary detail",fontdict=font1)
plt.xlabel('Id',fontdict=font2)
plt.ylabel('salary',fontdict=font2)
plt.show()
```
