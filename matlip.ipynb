```python
df=pd.read_csv('dunya.csv')
```

```python
df
```

```python
df['Country'].value_counts()
```

```python
plt.figure(figsize=(25,17))
plt.plot(df['Country'],df['Revenue'])
plt.xlabel('Country',fontsize=15)
plt.ylabel('Revenue',fontsize=15)
plt.show()
```

```python
dfx=pd.read_csv('ilk5.csv')
dfx
```

```python
fig, ax = plt.subplots(figsize=(30,5))  
ax.pie(df['Revenue'], labels=df['Country'], autopct='%1.1f%%', colors=['yellow', 'blue', 'green', 'red', 'purple','cyan','Magenta'])
ax.set_title('Revenue Distribution by Country', fontsize=16)

plt.show()
```

```python
fig.savefig("pie_Revenue.jpg",dpi=150)
```
